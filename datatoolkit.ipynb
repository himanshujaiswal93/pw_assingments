```python
#Q.14 Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
# b) Label the x-axis as 'Date'.
# c) Set the title of the plot as 'Temperature and Humidity Over Time'.


# Create a date range
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Create a DataFrame with random temperature and humidity data
np.random.seed(0)  # For reproducibility
temperature = np.random.uniform(low=15, high=35, size=100)  # Temperature between 15 and 35
humidity = np.random.uniform(low=30, high=90, size=100)  # Humidity between 30% and 90%

df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature
ax1.plot(df.index, df['Temperature'], color='red', label='Temperature', linewidth=2)
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df.index, df['Humidity'], color='blue', label='Humidity', linewidth=2)
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# b) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()

```

```python
# Q.15  Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
# a) Plot a histogram of the data with 30 bins.
# b) Overlay a line plot representing the normal distribution's probability density function (PDF).
#  c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
# d) Set the title of the plot as 'Histogram with PDF Overlay'.


from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay the normal distribution's probability density function (PDF)
mean = np.mean(data)
std_dev = np.std(data)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, color='red', linewidth=2, label='PDF')

# c) Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# q. 17
#  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import seaborn as sns


# Create two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.rand(100) * 20 - 10  # Random values between -10 and 10
y = np.random.rand(100) * 20 - 10  # Random values between -10 and 10

# Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Determine quadrants
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant I'
    elif x < 0 and y >= 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

df['Quadrant'] = df.apply(lambda row: get_quadrant(row['X'], row['Y']), axis=1)

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='Set1', style='Quadrant', s=100)

# Add legend, labels, and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')

# Show the plot
plt.grid()
plt.axhline(0, color='grey', lw=0.5, ls='--')  # Horizontal line at y=0
plt.axvline(0, color='grey', lw=0.5, ls='--')  # Vertical line at x=0
plt.show()


```

```python
# Q.18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'8

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output in a Jupyter Notebook (remove if running in a script)
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2π
y = np.sin(x)  # Sine of each x value

# Create a Bokeh figure
p = figure(title='Sine Wave Function', x_axis_label='X (radians)', y_axis_label='sin(X)', width=800, height=400)

# Add grid lines
p.grid.grid_line_alpha = 0.3  # Adjust grid line transparency

# Plot the sine wave
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')

# Show the plot
show(p)
```

```python
#Q,20  Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'


import plotly.graph_objs as go
import plotly.offline as pyo

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.rand(100)  # 100 random values

# Create a line plot
line_plot = go.Figure()

line_plot.add_trace(go.Scatter(
    x=x,
    y=y,
    mode='lines+markers',  # Show both lines and markers
    name='Random Data',
    line=dict(color='blue'),
    marker=dict(size=5)
))

# Label the axes
line_plot.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    template='plotly_white'  # Optional: use a white background
)

# Show the plot
pyo
```
